Se a apresentação for salva com o mesmo nome com que foi aberta, ela acrescentará as alterações nesta apresentação. Por outro lado, caso ela seja salva com um nome diferente, a apresentação que foi aberta não terá alteração. E uma nova apresentação será criada e salva, com as alteração, com o nove escolhido.

````python
replaces = {
    'Mês de Ano': f'{MESES[mes].title()} de {ano}',
    'MÊS ANO': f'{MESES[mes].upper()} {ano}',
    'JAN A MÊS': f'JAN A {MESES[mes].upper()}',
    'ACUMULADO ANO': f'ACUMULADO {ano}',
    'MÊS DE ANO (UPPER)': f'{MESES[mes].upper()} DE {ano}',
}
````

````python
def replace_text(replacements, shapes):
    """Takes dict of {match: replacement, ... } and replaces all matches.
    Currently not implemented for charts or graphics.
    """
    for shape in shapes:
        for match, replacement in replacements.items():
            if shape.has_text_frame:
                if (shape.text.find(match)) != -1:
                    text_frame = shape.text_frame
                    for paragraph in text_frame.paragraphs:
                        whole_text = "".join(run.text for run in paragraph.runs)
                        whole_text = whole_text.replace(str(match), str(replacement))
                        for idx, run in enumerate(paragraph.runs):
                            if idx != 0:
                                p = paragraph._p
                                p.remove(run._r)
                        if(not(not paragraph.runs)):
                            paragraph.runs[0].text = whole_text

if __name__ == '__main__':

    prs = Presentation('GDO - TESTE.pptx')
    # To get shapes in your slides
    slides = [slide for slide in prs.slides]
    shapes = []
    for slide in slides:
        for shape in slide.shapes:
            shapes.append(shape)

#     replaces = {
#                         '{{var1}}': 'text 1',
#                         '{{var2}}': 'text 2',
#                         '{{var3}}': 'text 3'
#                 }
    replace_text(replaces, shapes)
    prs.save('output.pptx')
````
